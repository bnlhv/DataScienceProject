```python
import pandas as pd
```

```python
table = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_Nintendo_Switch_video_games")
table = table[1].drop(['No.','As of','Release date[b]','Genre(s)',],axis=1)
table['Copies sold'] = table['Copies sold'].str.split('[',0)
table['Copies sold'] = table['Copies sold'].str[0]
table = table.dropna(axis=0, how = 'all')
table
```

```python

for i,text in enumerate(table['Copies sold']):
    if isinstance(text, str):
        if "million" in text:
            temp = text.replace('million' ,'')
            table['Copies sold'][i] = int((float(temp))*(1000000))
table
```
